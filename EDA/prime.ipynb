```python
import pandas as pd 
import numpy as np
import string
```

```python
df=pd.read_csv("raw_data/samples.csv")
```

```python
df
```

```python
unique_phenotypes=set()
for i in df["Phenotype"]:
    if(type(i)==float):
        continue
    i=i.lower()
    sep=i.split(",")
    for j in sep:
        unique_phenotypes.add(j)
```

```python
print(len(unique_phenotypes))
```

```python
for i in unique_phenotypes:
    print(i)
```

```python
yes_time_series=0
no_time_series=0
empty=0
daf=pd.DataFrame()

for i in range(len(df["Time_series"])):
    if(df["Time_series"][i].lower()=="yes"):
        yes_time_series+=1
        daf = pd.concat([daf, df.iloc[[i]]], ignore_index=True)
    elif(df["Time_series"][i].lower()=="no"):
        no_time_series+=1
    else:
        empty+=1
        
```

```python
print(yes_time_series)
print(no_time_series)
print(empty)
```
