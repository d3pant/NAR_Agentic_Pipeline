```python
import requests
import pandas as pd

def batch_query(accessions, batch_size=200):
    """Query ENA for BioSample IDs given a list of SRA run accessions."""
    results = []
    for i in range(0, len(accessions), batch_size):
        batch = accessions[i:i+batch_size]
        query = " OR ".join(f'run_accession="{a}"' for a in batch)
        url = "https://www.ebi.ac.uk/ena/portal/api/search"
        params = {
            "result": "read_run",
            "query": query,
            "fields": "run_accession,sample_accession,secondary_sample_accession,study_accession",
            "format": "tsv"
        }
        resp = requests.get(url, params=params)
        resp.raise_for_status()
        results.append(resp.text)
    return results
```

```python
# Take first 500 SRA run accessions and query ENA for their BioSample IDs
first_100_runs = df["Run"].iloc[:200].tolist()

raw_results = batch_query(first_100_runs, batch_size=200)

# Parse TSV response into a DataFrame
from io import StringIO

frames = []
for chunk in raw_results:
    lines = chunk.strip().split("\n")
    if len(lines) > 1:  # skip empty / header-only responses
        frames.append(pd.read_csv(StringIO(chunk), sep="\t"))

ena_df = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
print(f"ENA returned {len(ena_df)} rows for {len(first_100_runs)} queried runs")
ena_df.head()
```

```python
# Compare ENA-fetched BioSample IDs against the database (samples.csv) values
db_subset = df[["Run", "BioSample"]].iloc[:200].copy()
db_subset.columns = ["run_accession", "db_biosample"]

# Merge on run_accession; ENA returns sample_accession = BioSample ID
comparison = db_subset.merge(
    ena_df[["run_accession", "sample_accession"]].rename(columns={"sample_accession": "ena_biosample"}),
    on="run_accession",
    how="left"
)

comparison["match"] = comparison["db_biosample"] == comparison["ena_biosample"]

n_total     = len(comparison)
n_matched   = comparison["match"].sum()
n_mismatch  = (~comparison["match"] & comparison["ena_biosample"].notna()).sum()
n_not_found = comparison["ena_biosample"].isna().sum()

print(f"Total runs checked : {n_total}")
print(f"  Match            : {n_matched}  ({100*n_matched/n_total:.1f}%)")
print(f"  Mismatch         : {n_mismatch}  ({100*n_mismatch/n_total:.1f}%)")
print(f"  Not in ENA       : {n_not_found}  ({100*n_not_found/n_total:.1f}%)")

mismatches = comparison[~comparison["match"]]
if not mismatches.empty:
    print("\nMismatched / missing rows:")
    display(mismatches)
else:
    print("\nAll fetched BioSample IDs match the database.")
```

```python
import pandas as pd
import numpy as np

# --- config ---
ABUNDANCE_CSV_PATH = "raw_data/silva_species_absolute.csv"
TOP_N = 20
RANDOM_SEED = 42

# --- pick 10 random BioSamples that each have >1 run, keeping ALL runs ---
run_counts = df.groupby("BioSample")["Run"].count()
multi_run_biosamples = run_counts[run_counts > 1].index.tolist()

rng = np.random.default_rng(RANDOM_SEED)
selected_biosamples = rng.choice(multi_run_biosamples, size=10, replace=False).tolist()

sample_map = (
    df[df["BioSample"].isin(selected_biosamples)][["BioSample", "Run"]]
    .dropna()
    .reset_index(drop=True)
)

print(f"Selected 10 BioSamples → {len(sample_map)} total runs:")
display(sample_map.groupby("BioSample")["Run"].apply(list).reset_index().rename(columns={"Run": "runs"}))

target_runs = sample_map["Run"].tolist()
biosample_of = dict(zip(sample_map["Run"], sample_map["BioSample"]))
```

```python
# --- load only the relevant run columns from the abundance CSV ---
# Raw format: rows = Run IDs (first column), remaining columns = taxa lineage strings
# We read in chunks and keep only our 10 runs to avoid loading the full ~53k-column matrix

abund_raw = pd.read_csv(ABUNDANCE_CSV_PATH, index_col=0)

missing_runs = [r for r in target_runs if r not in abund_raw.index]
if missing_runs:
    print(f"Warning: {len(missing_runs)} run(s) not found in abundance CSV: {missing_runs}")

found_runs = [r for r in target_runs if r in abund_raw.index]
abund_subset = abund_raw.loc[found_runs]   # shape: (found_runs, taxa)

print(f"Abundance matrix slice: {abund_subset.shape[0]} runs × {abund_subset.shape[1]} taxa")
```

```python
# --- show top 20 taxa per run by absolute abundance, with relative abundance alongside ---
for run in found_runs:
    biosample = biosample_of[run]
    row = abund_subset.loc[run]
    total_reads = row.sum()

    top20 = (
        row.sort_values(ascending=False)
        .head(TOP_N)
        .to_frame(name="absolute_abundance")
    )
    top20.index.name = "taxon"
    top20 = top20[top20["absolute_abundance"] > 0].copy()
    top20["relative_abundance"] = (top20["absolute_abundance"] / total_reads).round(4)

    print(f"\n{'='*70}")
    print(f"BioSample : {biosample}")
    print(f"Run       : {run}")
    print(f"Total reads: {int(total_reads):,}")
    print(f"{'='*70}")
    display(top20)
```
